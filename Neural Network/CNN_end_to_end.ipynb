```python
# print(arr.shape[1]-1)
# arr = np.vstack((arr,arr))
X = np.array(arr[:,0:arr.shape[1]-1])
print("The feature in X are-:", X.shape[1])
Y = np.array(arr[:, arr.shape[1]-1].reshape([-1,1]), dtype = int)
print("The shape of Y is-:", Y.shape)
num_classes = int(np.max(Y))+1
print("Total number of classes are-: ", num_classes)
train_x = None
val_x = None
train_y = None
val_y = None
for i in range(num_classes):
    indices = np.where(Y==i)
    print(indices)
    train_id = int(indices[0][0]+0.8*len(indices[0]))
    print(indices[0][0])
    print(indices[0][-1])
#     print(len(indices))
    x_train = X[indices[0][0]:train_id, :]
    x_val = X[train_id:indices[0][-1], :]
    y_train = Y[indices[0][0]:train_id].reshape([-1,1])
    y_val = Y[train_id:indices[0][-1]].reshape([-1,1])
#     print(y_train.shape[0])
#     print(y_val.shape[0],y_val.shape[1])
    if i==0:
        train_x = x_train
        train_y = y_train
        val_x = x_val
        val_y = y_val
    else : 
        train_x = np.vstack((train_x, x_train))
        train_y = np.vstack((train_y, y_train))
        val_y = np.vstack((val_y, y_val))
        val_x = np.vstack((val_x, x_val))
#     print(train_x)
#     print(train_y)
print(val_x.shape)
print(train_x.shape)
print(val_y.shape)
print(train_y.shape)

# print(np.min(Y))
# print(np.max(Y))
print(train_y[3450][0])
y_one_hot_train = np.zeros((train_x.shape[0], num_classes))
y_one_hot_val = np.zeros((val_x.shape[0], num_classes))

for i in range(train_y.shape[0]):
    y_one_hot_train[i, train_y[i][0]] = 1
for i in range(val_y.shape[0]):
    y_one_hot_val[i, val_y[i][0]] = 1
print(y_one_hot_train)
```

```python
WIDTH = width
num_features = int(X.shape[1])
HEIGHT = height

train_x = train_x.reshape([-1,WIDTH,HEIGHT,3])
val_x = val_x.reshape([-1,WIDTH,HEIGHT,3])
print(train_x.shape)
print(val_x.shape)

def net(x):
    conv1 = tf.nn.dropout(tf.layers.conv2d(x,8,kernel_size = [3,3], strides= [1,1], padding= 'same'), keep_prob = 0.8)
    pool1 = tf.layers.max_pooling2d(conv1,pool_size=[2,2], strides = [2,2])
    a1 = tf.nn.relu(pool1)
    conv2 = tf.nn.dropout(tf.layers.conv2d(a1,16,kernel_size = [3,3], strides= [1,1], padding= 'same'), keep_prob = 0.8)
    pool2 = tf.layers.max_pooling2d(conv2,pool_size=[2,2], strides = [2,2])
    a2 = tf.nn.relu(pool2)
    
#     conv3 = tf.nn.dropout(tf.nn.conv2D(x,8,kernel_size = [3,3], strides= [1,1], padding= 'same'), keep_prob = 0.8)
#     pool3 = tf.layers.maxpooling2d(conv1,pool_size=[2,2], strides = [2,2])
#     a3 = tf.nn.relu(pool1)
#     conv4 = tf.nn.dropout(tf.nn.conv2D(x,8,kernel_size = [3,3], strides= [1,1], padding= 'same'), keep_prob = 0.8)
#     pool4 = tf.layers.maxpooling2d(conv1,pool_size=[2,2], strides = [2,2])
#     a4 = tf.nn.relu(pool1)
    
    flat3 = tf.contrib.layers.flatten(pool2)
    full3 = tf.layers.dense(flat3,400)
    a3 = tf.nn.relu(full3)
        
    full4 = tf.layers.dense(a3,num_classes)
    probs = tf.nn.sigmoid(full4)
    
    return probs
```

```python
num_epochs = 1
batch_size = 128
lr = 1e-3
sess = tf.Session()

X_tf = tf.placeholder(shape=[None,WIDTH,HEIGHT,3],name='X_tf',dtype=tf.float32)
Y_tf = tf.placeholder(shape=[None,num_classes],name='Y_tf',dtype=tf.float32)

probs= net(X_tf)
y_hat = tf.argmax(probs,axis=1)
y_compare = tf.argmax(Y_tf,axis=1)
accuracy_ = tf.reduce_mean(tf.cast(tf.equal(y_hat,y_compare),tf.float32))
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=probs, labels=Y_tf))
trainer = tf.train.AdamOptimizer(lr)
train_step = trainer.minimize(loss)

for epoch in range(num_epochs):
    print('Epoch-: ' + str(epoch) )
    
    perms = np.random.permutation(train_x.shape[0])
    num_batches = (int)(train_x.shape[0]/batch_size)

    X_shuffle = train_x[perms,:]
    y_shuffle = y_one_hot_train[perms,:]
    
    epoch_loss = 0.0
    epoch_acc = 0.0
    epoch_val_acc = 0.0
    
    for i in range(num_batches):
#         print(i)
        X_batch = X_shuffle[i*batch_size:(i+1)*batch_size,:]
        y_batch = y_shuffle[i*batch_size:(i+1)*batch_size,:]
        
        l,acc = sess.run([train_step,loss,accuracy_],feed_dict={X_tf:X_batch,Y_tf:y_batch})
        # print('it : ' + str(it))
        epoch_loss+=l
        epoch_acc+=acc
    
    perms = np.random.permutation(val_x.shape[0])

    X_val_shuffle = val_x[perms,:]
    y_val_shuffle = y_one_hot_val[perms,:]
    
    X_val_batch = X_val_shuffle[0:batch_size,:]
    y_val_batch = y_val_shuffle[0:batch_size,:]
    
    val_acc_list = sess.run([accuracy_],feed_dict={X_tf:X_val_batch,y_tf:y_val_batch})    
    val_acc = val_acc_list
    # print('it : ' + str(it))
    
    print('Learning reate-: : ' + str(lr))
    loss_vector.append(epoch_loss/(num_batches*1.0))
    train_acc_vector.append(epoch_acc/(num_batches*1.0))
    val_acc_vector.append(val_acc)
    
    if epoch%5 == 0:
        print('Epoch = '+ str(epoch))
        print('Loss = ' + str(epoch_loss/(num_batches*1.0)))
        print('Accuracy = ' + str(epoch_acc/(num_batches*1.0)))
        print('Validation Accuracy = ' + str(val_acc))
```
