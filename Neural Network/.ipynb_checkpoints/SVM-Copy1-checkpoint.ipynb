```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn import datasets
from tensorflow.python.framework import ops
ops.reset_default_graph()
```

```python
iris = datasets.load_iris()
x_vals = np.array([[x[0], x[3]] for x in iris.data])

X_data,y_data = make_circles(n_samples=450,factor=.3,noise=.1)
y_data[y_data == 0] = -1
X_train = (X_data[:300,:]).reshape([-1,2])
y_train = (y_data[:300]).reshape([-1,1])
X_val = (X_data[300:400,:]).reshape([-1,2])
y_val = (y_data[300:400]).reshape([-1,1])
X_test = (X_data[400:,:]).reshape([-1,2])
y_test = (y_data[400:]).reshape([-1,1])

print(y_data)
plt.scatter(X_data[:,0],X_data[:,1],c=y_data)
plt.show()
```

```python
batch_size = 100
gamma = 50.0

tf.reset_default_graph()
X_batch = tf.placeholder(name='X_batch',shape=[None,2],dtype=tf.float32)
y_batch = tf.placeholder(name='y_batch',shape=[None,1],dtype=tf.float32)
X_prediction_support = tf.placeholder(name='X_prediction_support',shape=[None,2],dtype=tf.float32)
y_prediction_support = tf.placeholder(name='y_prediction_support',shape=[None,1],dtype=tf.float32) #Used in prediction

X_prediction = tf.placeholder(name='X_prediction',shape=[None,2],dtype=tf.float32)
y_target = tf.placeholder(name='y_target',shape=[None,1],dtype=tf.float32) #used in accuracy 

alpha = tf.Variable(tf.random_normal(shape=[1,batch_size]))
```

```python
#Kernel
#example - exp(-gamma*(||x1 - x2||**2))
'''x - [[x1 y1]
        [x2 y2]]
'''    
squared_distance_reduced = tf.reshape(tf.reduce_sum(tf.square(X_batch),axis=1),[-1,1])
print(squared_distance_reduced.shape)
'''
out : [[x1**2 + y1**2]
       [x2**2 + y2**2]]
'''
correlation_mat = tf.matmul(X_batch,tf.transpose(X_batch)) #m,m
'''
out : [[x1**2+y1**2  x1x2 + y1y2]
       [x1x2+y1y2    x2**2+y2**2]]
'''
print(correlation_mat.shape)
exponential_power = squared_distance_reduced - 2.0*correlation_mat + tf.transpose(squared_distance_reduced)
'''
out : ||x1 - x2||**2
'''
kernel = tf.exp(-1.0*gamma*exponential_power) #Rbf kernel
print(kernel.shape)
```

```python
alpha_correlation = tf.matmul(tf.transpose(alpha),alpha)
y_correlation = tf.matmul(y_batch,tf.transpose(y_batch))

loss_second_term = tf.reduce_sum(alpha_correlation*y_correlation*kernel)/2.0
loss_first_term = tf.reduce_sum(alpha)

loss = loss_second_term - loss_first_term
```

```python
#kernel for testing
squared_support_vectors = tf.reshape(tf.reduce_sum(tf.square(X_prediction_support),axis=1),[-1,1])
squared_prediction_vectors = tf.reshape(tf.reduce_sum(tf.square(X_prediction),axis=1),[-1,1])

correlation_mat_test = tf.matmul(X_prediction_support,tf.transpose(X_prediction))

exponential_power_test = squared_support_vectors - 2.0*correlation_mat_test + tf.transpose(squared_prediction_vectors)

kernel_test = tf.exp(-1.0*gamma*exponential_power_test)
```

```python
#accuracy
prediction_output = tf.matmul(tf.multiply(tf.transpose(y_prediction_support),alpha),kernel_test)
prediction = tf.sign(prediction_output-tf.reduce_mean(prediction_output))
accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.squeeze(prediction), tf.squeeze(y_target)), tf.float32))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.002)
train_step = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)

num_epochs = 1000
for epoch in range(num_epochs):
    rand_index = np.random.choice(300, size=batch_size)
    rand_x = X_train[rand_index]
    rand_y = [y_train[rand_index]]
    rand_y = np.array(rand_y).reshape([-1,1])
    
    rand_index_i = np.random.choice(300,size=100) #for test
    rand_x_support = X_train[rand_index_i]
    rand_y_support = [y_train[rand_index_i]]
    rand_y_support = np.array(rand_y_support).reshape([-1,1])

    _,l,acc = sess.run([train_step,loss,accuracy],feed_dict={X_batch:rand_x,y_batch:rand_y,X_prediction_support:rand_x_support,y_prediction_support:rand_y_support,X_prediction:X_val,y_target:y_val})
    
    print(l)
    print('Accuracy : ' + str(acc))
    
sess.close()
```
