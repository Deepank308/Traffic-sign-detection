```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Dataset generation reference : 
N = 100 # number of points per class
D = 2 # dimensionality
K = 3 # number of classes
X = np.zeros((N*K,D)) # data matrix (each row = single example)
y = np.zeros(N*K, dtype='uint8') # class labels
for j in range(K):
  ix = range(N*j,N*(j+1))
  r = np.linspace(0.0,1,N) # radius
  t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
  X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
  y[ix] = j
# lets visualize the data:
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.show()

#One-hot encode 
y_o = np.zeros((N*K,K))
for i in range(N*K):
    y_o[i,y[i]] = 1
print(y_o.shape)
```

```python
class NeuralNet:
    def __init__(self,reg):
        self.reg = reg
        self.units = []
        self.activations = []
    
    def add_layer(self,num_units,activation='relu'): #units = [3],act = ['relu']
        self.activations.append(activation)
        self.units.append(num_units)
    
    def compile_net(self,X,y): #y - one-hot-encoded
        self.weights = []
        self.biases = []
        # print(X.shape)
        for unit_id in range(len(self.units)): #range(1)
            if unit_id == 0:
                W = tf.get_variable(name='W'+str(unit_id),shape=[X.shape[1],self.units[unit_id]],initializer=tf.contrib.layers.xavier_initializer())
                b = tf.get_variable(name='b'+str(unit_id),shape=[1,self.units[unit_id]],initializer=tf.zeros_initializer())
                self.weights.append(W)
                self.biases.append(b)
            else:
                W = tf.get_variable(name='W'+str(unit_id),shape=[self.units[unit_id-1],self.units[unit_id]],initializer=tf.contrib.layers.xavier_initializer())
                b = tf.get_variable(name='b'+str(unit_id),shape=[1,self.units[unit_id]],initializer=tf.zeros_initializer())
                self.weights.append(W)
                self.biases.append(b)
        
        # print(y.shape)
        W = tf.get_variable(name='W'+str(len(self.units)),shape=[self.units[-1],y.shape[1]],initializer=tf.contrib.layers.xavier_initializer())
        b = tf.get_variable(name='b'+str(len(self.units)),shape=[1,y.shape[1]],initializer=tf.zeros_initializer())
        self.weights.append(W)
        self.biases.append(b)
        
        self.activations.append('') #TODO : Make it softmax
        
    def execute(self,X,y):
        activations = []
        activations.append(X)
        
        counter = 0
        for W,b in zip(self.weights,self.biases):
            A = 0
            if counter == 0:
                # print('counter = 0 : ')
                A = tf.matmul(X,W) + b
            else:
                # print(activations[counter].shape)
                # print(W.shape)
                A = tf.matmul(activations[counter],W) + b
            A_out = 0
            if self.activations[counter] == 'relu':
                # print('relu')
                A_out = tf.nn.relu(A)
            elif self.activations[counter] == 'sigmoid':
                # print('sigmoid')
                A_out = tf.nn.sigmoid(A)
            elif self.activations[counter] == 'tanh':
                # print('tanh')
                A_out = tf.nn.tanh(A)
            else:
                A_out = A
            counter+=1
            activations.append(A_out)
            
        c_ = tf.cast(tf.argmax(activations[-1],axis=1),dtype=tf.float32)
        y_acc = tf.cast(tf.argmax(y,axis=1),dtype=tf.float32)
        acc = tf.reduce_mean(tf.constant(1.0) - tf.abs(c_-y_acc))*100.0
        
        loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=A_out,labels=y))
        
        for W in self.weights:
            loss+=(tf.reduce_sum(W*W))*self.reg
        
        trainer = tf.train.AdamOptimizer(learning_rate=1e-3).minimize(loss) #check for now
        # print('t')
        return [loss,trainer,acc]
        
    def predict(self,X):
        activations = []
        activations.append(X)
        
        counter = 0
        for W,b in zip(self.weights,self.biases):
            A = 0
            if counter == 0:
                # print('counter = 0 : ')
                A = tf.matmul(X,W) + b
            else:
                # print(activations[counter].shape)
                # print(W.shape)
                A = tf.matmul(activations[counter],W) + b
            A_out = 0
            if self.activations[counter] == 'relu':
                # print('relu')
                A_out = tf.nn.relu(A)
            elif self.activations[counter] == 'sigmoid':
                # print('sigmoid')
                A_out = tf.nn.sigmoid(A)
            elif self.activations[counter] == 'tanh':
                # print('tanh')
                A_out = tf.nn.tanh(A)
            else:
                A_out = A
            counter+=1
            activations.append(A_out)
        
        c_ = tf.argmax(activations[-1],axis=1)
        
        return c_
```

```python
tf.reset_default_graph()
net = NeuralNet(0.0)
net.add_layer(3,'sigmoid')
X_tf = tf.placeholder(name='X_tf',shape=[None,D],dtype=tf.float32)
y_tf = tf.placeholder(name='y_tf',shape=[None,K],dtype=tf.float32)
sess = tf.Session()
net.compile_net(X_tf,y_tf)
loss,trainer,accuracy = net.execute(X_tf,y_tf)
init_g = tf.global_variables_initializer()
init_l = tf.local_variables_initializer()
sess.run(init_g)
sess.run(init_l)
for epoch in range(100000):
    los,_,acc = sess.run([loss,trainer,accuracy],feed_dict={X_tf:X,y_tf:y_o})
    # print("C_ : " + str(c_t))
    # print("Y_acct : " + str(y_acct))
    
    if epoch%1000 == 0:
        print(los)
        print("Accuracy : " + str(acc))
        
        if epoch%10000 == 0:
            pass
            # plot_decision_boundary(X,y)
```

```python
def plot_decision_boundary(X,y):
    x_min = X[:,0].min()
    x_max = X[:,0].max()
    y_min = X[:,1].min()
    y_max = X[:,1].max()
    
#     print(x_min)
#     print(x_max)
    X_plot = np.copy(X)
    y_plot = np.copy(y)
    # print(y_plot)
    h = 0.2
    xs = np.linspace(x_min,x_max,int((x_max-x_min)/h))
    ys = np.linspace(y_min,y_max,int((y_max-y_min)/h))
    
    X_t = tf.placeholder(name='X_pred',shape=[None,D],dtype=tf.float32)

    for x in xs:
        for y in ys:
            xten = np.array([x,y]).reshape(1,2)
            # xten = tf.convert_to_tensor(xten,dtype=tf.float32)
            c = net.predict(X_t)
            pred = sess.run(c,feed_dict={X_t:xten})
            # print(pred)
            pred = np.array(pred)
            # print(pred)
            X_plot = np.vstack([X_plot,np.array([x,y])])
            y_plot = np.hstack([y_plot,pred])

    plt.scatter(X_plot[:,0],X_plot[:,1],c=y_plot)
    plt.show()

plot_decision_boundary(X,y)

sess.close()
```
